```python
import os
import matplotlib.pyplot as plt

def count_files_in_folders(base_folder):
    folder_counts = {}
    
    for root, dirs, files in os.walk(base_folder):
        for folder in dirs:
            folder_path = os.path.join(root, folder)
            file_count = len(os.listdir(folder_path))
            folder_counts[folder] = file_count
    
    return folder_counts

# Thay đổi đường dẫn này thành đường dẫn tới folder của bạn
base_folder = "./"
folder_counts = count_files_in_folders(base_folder)

# Vẽ biểu đồ cột ngang
folders = list(folder_counts.keys())
counts = list(folder_counts.values())
for i in range(len(folders)):
    print(f"{folders[i]}: {counts[i]} files")
print(sum(counts), "files in total")
plt.barh(folders, counts, color='skyblue')
plt.xlabel('Số lượng file')
plt.title('Các domain')
plt.show()
```
